```python
import torch
from PIL import Image
from lbm.inference import evaluate, get_model

# Load model
model = get_model(
    # "jasperai/LBM_depth",
    "jasperai/LBM_normals",
    torch_dtype=torch.bfloat16,
    device="cuda",
)
```

```python
# Load a source image
source_image = Image.open("LBM/assets/photo_2025-09-16_17-45-38.jpg")

# Perform inference
output_image = evaluate(model, source_image, num_sampling_steps=2)
```
